Програма приймає на вхід два параметри: стать спортсмена(`sex = (f, m)`) та вік спортсмена(`age = [0,100]`).  Спортсмен-дівчина може брати участь у змаганні, якщо її вік більший 18, спортсмен-хлопець - якщо вік більший 16

Якщо введеніть не коректні дані, виведіть повідомлення про помилку

```python
sex = input()
age = float(input())

if sex == 'f' and age > 18:
    print('Good')
elif sex == 'f' and age < 18:
    print('too young')        
elif sex == 'm' and age > 16:
    print('Good')
elif sex == 'm' and age < 16:
    print('too young')        
else:
    print('Incorrect values')
```

### Завдання 6 

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random

#генеруэмо тисячу чисел в діапазоні від 0 до 10
random_numbers = [random.randint(0,100) for i in range(1000)]

# рохраховує мінімальне та максимальне значення
print('1. мін =', min(random_numbers), 'макс =', max(random_numbers))

# середнє значення послідовності
ariphmetic_mean = sum(random_numbers) / len(random_numbers)
print('2. середнє значення =', ariphmetic_mean)

# медіану послідовності
if len(random_numbers) % 2 == 0:
    right_index = len(random_numbers) // 2
    median = (random_numbers[right_index - 1] + random_numbers[right_index]) / 2
else:
    median = random_numbers[len(random_numbers) // 2]
    
print('3. медіана =', median)

# частоту кожної цифри в послідовності
print('4. частота цифр:\n', {num:random_numbers.count(num) for num in set(random_numbers)})

# середнє квадратичне відхилення
result = 0
for n in random_numbers:
    result += (n - ariphmetic_mean) ** 2

print('5. середнє квадратичне відхилення =', (result / len(random_numbers)) ** 0.5)
```
