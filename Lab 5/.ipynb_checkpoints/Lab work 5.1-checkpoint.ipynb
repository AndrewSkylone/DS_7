Заповнити вектор з 20 елементів випадковими цілими числами. Знайти індекси ненульових елементів:

```python
vec = np.random.randint(0, 10, 20)
print(vec)
vec.nonzero()
```

Для вектора із попередньої задачі знайти середнє значення, медіану, стандартне відхилення:

```python
print(f'середнє знач = {np.mean(vec)}')
print(f'медіана = {np.median(vec)}')
print(f'стандартне відхилення = {np.std(vec)}')
```

Створити дві матриці випадкових чисел розміром 4х2 та 2х2. Перемножити їх:

```python
A1 = np.random.random((4, 2))
A2 = np.random.random((2, 2))
A = A1 * A2[:, np.newaxis]
A
```

Результуючу матрицю із попередньої задачі перетворити у вектор. Знайти мінімальне та максимальне значення. 

Максимальне значення замінити на 0:

```python
vec = A.reshape(-1)

print('max =', vec.max())
print('min =', vec.min())

max_ind = np.where(vec == vec.max())
vec[max_ind] = 0
print('vec[max_ind]: ', vec[max_ind])
```

Знайдіть кореляцію першого рядка матриці із попереднього завдання із іншими рядками:

```python
A = A.reshape(8, 2)
np.apply_along_axis(lambda x: np.correlate(A[1], x), axis=1, arr=A[1:])
```

Дано матрицю. Знайти ранг матриці:

```python
a = np.array([[ 1,  4, 5],
              [7, 3, 2],
              [9, 16, 1]])
print('ранг матриці = ', np.linalg.matrix_rank(a))
```

Знайти визначник матриці:

```python
print('визначник матриці = ', np.linalg.det(a))
```

Знайти власні значення та власні вектори матриці: 

```python
print('власні значення = ', np.linalg.eig(a)[0])
print('власні вектори = ', np.linalg.eig(a)[1])
```

Знайдіть число, що найчастіше зустрічається в масиві:

```python
a = np.random.randint(0,10,50)
np.bincount(a).argmax()
```

Виведіть лише парні числа, що діляться на 5:

```python
a = np.random.randint(0,10,50)
a[(a % 2 == 0) & (a % 5 == 0) & (a != 0)]
```

Розв'яжіть систему рівнянь:
$$2x + 5y = 1 $$

$$x - 10y = 3  $$
   

```python
M = np.array([[2., 5.], [1., -10.]])
v = np.array([1., 3.])

np.linalg.solve(M, v)
```
