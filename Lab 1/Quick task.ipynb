2. Напишіть функцію, що приймає координати двох точок та виводить відстань між ними

```python
x1 = int(input('x1: '))
y1 = int(input('y1: '))
x2 = int(input('x2: '))
y2 = int(input('y2: '))

((x1 - x2)**2 + (y1 - y2)**2)**0.5
```

3. Дана послідовність натуральних чисел. Відфільтруйте та виведіть лише прості числа

```python
N = range(1, 60)
simple = []

for natural in N:
    if all(natural % i != 0 for i in range(2, natural)):
        simple.append(natural)
simple    
```

4. Дано список, елементи якого є слова. Змініть літеру кожного слова на велику та виведіть отриманий список

```python
L = ['Hello', 'world', 'and', 'his', 'residents']
[s.title() for s in L]
```
